## 홈페이지 들어갈 제주도 정보 관련 지도 메이킹

```python
from folium.features import CustomIcon
```

```python
m = folium.Map(
    location=[33.36467199334479, 126.52865102552187],
    zoom_start = 10,
    min_zoom = 10, 
    max_zoom = 25,
)

tooltip = 'Click!'
# filter기능으로 서귀포 오름 보기, 카페 정보, 맛집 정보, 해수욕장 정보
오름 = folium.FeatureGroup(name='서귀포오름').add_to(m)
카페 = folium.FeatureGroup(name='카페모음').add_to(m)
맛집 = folium.FeatureGroup(name='맛집모음').add_to(m)
해수욕장 = folium.FeatureGroup(name='해수욕장모음').add_to(m)

# 오름 icon설정(산 이미지 다운)
for i in range(오름.shape[0]):
  if {오름}
  icon_image = 'mountain.png'
  icon = CustomIcon(
      icon_image,
      icon_size = (20,20),
      popup_anchor = (30, -30)
  )
  folium.Marker(
      [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
      popup = f"<div style='width:100px'><strong>{오름.iloc[i]['오름명']}</strong><br>\
      화장실 : {오름.iloc[i]['화장실']}<br>\
      주차장 : {오름.iloc[i]['주차장']}<br>\
      <a href='https://www.visitjeju.net/kr/search?q=%EC%98%A4%EB%A6%84#'>정보 더보기</a></div>",
      tooltip = tooltip, 
      icon = icon
  ).add_to(오름)

m
```
